```python
# 'country', 'region', 'rank', 'score', 'GDP', 'social', 'healthy', 'freedom', 'generosity', 'corruption'
happy2015 = happy2015[['Country', 'Region', 'Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Generosity', 'Trust (Government Corruption)']]
happy2015.columns = new_columns
happy2015
```

```python
# 'country', 'region', 'rank', 'score', 'GDP', 'social', 'healthy', 'freedom', 'generosity', 'corruption'
happy2016 = happy2016[['Country', 'Region', 'Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Generosity', 'Trust (Government Corruption)']]
happy2016.columns = new_columns
happy2016
```

```python
# 'country', 'region', 'rank', 'score', 'GDP', 'social', 'healthy', 'freedom', 'generosity', 'corruption'
happy2017['region'] = pd.Series([])
happy2017 = happy2017[['Country', 'region', 'Happiness.Rank', 'Happiness.Score', 'Economy..GDP.per.Capita.', 'Family', 'Health..Life.Expectancy.', 'Freedom', 'Generosity', 'Trust..Government.Corruption.']]
happy2017.columns = new_columns
happy2017
```

```python
# add region from previous dataset and check
list_of_countries = happy2017['country']
for i in list_of_countries:
    a = ''.join(happy2016[happy2016['country']==i]['region'].values)
    b = happy2017[happy2017['country']==i].index
    happy2017.loc[b, 'region'] = a
happy2017.info()
```

```python
# 'country', 'region', 'rank', 'score', 'GDP', 'social', 'healthy', 'freedom', 'generosity', 'corruption'
happy2018['region'] = pd.Series([])
happy2018 = happy2018[['Country or region', 'region', 'Overall rank', 'Score', 'GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']]
happy2018.columns = new_columns
happy2018
```

```python
# add region from previous dataset and check
# add 0 into NaN cell 'corruption' in row 19
list_of_countries = happy2018['country']
for i in list_of_countries:
    a = ''.join(happy2016[happy2016['country']==i]['region'].values)
    b = happy2018[happy2018['country']==i].index
    happy2018.loc[b, 'region'] = a
happy2018.loc[19, 'corruption'] = 0
happy2018.info()
```

```python
# 'country', 'region', 'rank', 'score', 'GDP', 'social', 'healthy', 'freedom', 'generosity', 'corruption'
happy2019['region'] = pd.Series([])
happy2019 = happy2019[['Country or region', 'region', 'Overall rank', 'Score', 'GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']]
happy2019.columns = new_columns
happy2019
```

```python
# add region from previous dataset and check
list_of_countries = happy2019['country']
for i in list_of_countries:
    a = ''.join(happy2016[happy2016['country']==i]['region'].values)
    b = happy2019[happy2019['country']==i].index
    happy2019.loc[b, 'region'] = a
happy2019.info()
```

```python
# we can add year into each dataset for next researching
# and concate all datasets into one which named 'happy'
happy2015['year'] = 2015
happy2016['year'] = 2016
happy2017['year'] = 2017
happy2018['year'] = 2018
happy2019['year'] = 2019
happy = pd.concat([happy2015, happy2016, happy2017, happy2018, happy2019], ignore_index=True)
happy
```

```python
happy_top_5_2015 = happy2015.head(5)
happy_top_5_2016 = happy2016.head(5)
happy_top_5_2017 = happy2017.head(5)
happy_top_5_2018 = happy2018.head(5)
happy_top_5_2019 = happy2019.head(5)
happy_top_5 = pd.concat([happy_top_5_2015 , happy_top_5_2016, happy_top_5_2017, happy_top_5_2018, happy_top_5_2019], ignore_index=True)
```

```python
happy_top_5.groupby('country')['country'].count().sort_values(ascending=False)
```

```python
happiest_country = happy_top_5[['country', 'score', 'GDP', 'social', 'healthy', 'freedom', 'generosity', 'corruption']]
data = happiest_country.groupby('country').mean().sort_values(['score'], ascending=False)
```

```python
sns.lineplot(data=data, x='score', y='GDP')
```
